```python
import os
from collections import namedtuple

# pip install Pillow
from PIL import Image


Record = namedtuple(
    'Record',
    ['host', 'screenshot', 'markers']
)


def load_lines(path):
    with open(path) as file:
        for line in file:
            yield line.rstrip('\n')


def load_data(dir):
    screenshots = {}
    host_markers = {}
    for filename in os.listdir(dir):
        host, extension = os.path.splitext(filename)
        path = os.path.join(dir, filename)
        if extension == '.txt':
            markers = list(load_lines(path))
            host_markers[host] = markers
        elif extension == '.png':
            screenshot = Image.open(path)
            screenshots[host] = screenshot
    for host in screenshots:
        screenshot = screenshots[host]
        markers = host_markers[host]
        yield Record(host, screenshot, markers)

        
data = sorted(load_data('data'))
```

```python
from IPython.display import display


def display_record(record):
    display(record.screenshot)
    print('http://' + record.host)
    print(record.markers)
    
    
display_record(data[91])
```

```python
from ipyannotate.buttons import (
    ValueButton as Button,
    NextButton as Next,
    BackButton as Back
)
from ipyannotate.toolbar import Toolbar
from ipyannotate.tasks import Task, Tasks
from ipyannotate.canvas import OutputCanvas
from ipyannotate.annotation import Annotation


tasks = Tasks(Task(_) for _ in data)
buttons = [
    Button('sexshop', shortcut='1'),
    Button('porn', shortcut='2'),
    Button('pills', shortcut='3'),
    Button('striptease', shortcut='4'),
    Button('massage', shortcut='5'),
    Button('dating', shortcut='6'),
    Button('hookers', shortcut='7'),
    Button('club'),
    Button('gay'),
    Button('news'),
    Button('fetish'),
    Button('stories'),
    Button('games'),
    Button('forum'),
]
controls = [
    Back(),
    Next()
]
toolbar = Toolbar(buttons + controls)
canvas = OutputCanvas(display=display_record)
annotation = Annotation(toolbar, tasks, canvas=canvas)
annotation
```
