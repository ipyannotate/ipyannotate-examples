```python
import os
from collections import namedtuple

# pip install Pillow
from PIL import Image


Screenshot = namedtuple('Screenshot', ['host', 'image'])
        

def load_screenshots(dir):
    for filename in os.listdir(dir):
        host, extension = os.path.splitext(filename)
        if extension == '.png':
            path = os.path.join(dir, filename)
            image = Image.open(path)
            yield Screenshot(host, image)

            
screenshots = list(load_screenshots('screenshots/'))
```

```python
from IPython.display import display


def show_screenshot(screenshot):
    print('http://' + screenshot.host)
    display(screenshot.image)
    

for screenshot in screenshots[:3]:
    show_screenshot(screenshot)
```

```python
from ipyannotate import annotate

annotation = annotate(screenshots, display=show_screenshot)
ok, error, next, back = annotation.toolbar.buttons
error.label = 'parked'
annotation
```
